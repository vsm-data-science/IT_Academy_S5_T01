## Exercici 3

Descarrega el dataset adjunt que tracta de les diferents peticions ciutadanes que gestiona l'Ajuntament de Barcelona. Amb la funció GroupBy, agrupa el dataset per la variable 'Tipus' i realitza un recompte (count) per tal de poder mostrar en un gràfic de barres la quantitat de gestions de cada tipus que fa l'Ajuntament de Barcelona

```python
#primer pujo el document csv
peti_ciutat = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Peticions ciutadanes_OpenDataBCN.csv")

print(peti_ciutat)
```

```python
# Faig un recompte de cada valor únic que té la 'TIPUS'
recompte_tipus = peti_ciutat.groupby('TIPUS')['TIPUS'].count()


# i a continuació començo a fer el gráfic de barres
plt.figure(figsize=(7, 6))  # aquest es el tamany del gráfic
recompte_tipus.plot(kind='bar', color='blue')  # li dic que el vull de barres i que siguin de color blau
plt.title("Quantitat de gestions de cada tipus que fa l'Ajuntament de Barcelona") # títol
plt.xlabel('Tipus de gestió')  # Etiqueta eix X
plt.ylabel('Número de gestions')  # Etiqueta eix Y
plt.xticks(rotation=45)  # inclinació de les etiquetes per que es poguin llegir millor, posarem 45º per que quedin inclinades.

plt.show()
```

## Exercici 4

Realitza una visualització gràfica de dues variables (TIPUS, MES_DATA_TANCAMENT) que permeti visualitzar clarament la informació de les dues variables (tingues en compte que el dataset només té registres dels 2 primers mesos del 2022).



```python

# Calcular el recuento de incidencias por tipo y mes
count_data = peti_ciutat.groupby(['TIPUS', 'MES_DATA_TANCAMENT']).size().unstack().fillna(0)

# Crear la gráfica de barras
count_data.plot(kind='bar')

# Configurar etiquetas y leyendas
plt.xlabel("Tipus d'incidencies")
plt.ylabel("num. d'incidencies tencades")
plt.title("Recompte del número d'incidencies tencades per mes")

plt.legend(title='Mes de tancament', labels=['Gener', 'Febrer'])

# Mostrar la gráfica
plt.show()

```

Realitza una visualització gràfica del dataset adjunt amb alguna informació que consideris rellevant mostrar per mitjà d'un gràfic. Pots usar tantes variables com vulguis.

```python
recompte_area = peti_ciutat['AREA'].value_counts()

# Creem el gráfic
plt.figure(figsize=(20, 6))
recompte_area.plot(kind='bar', color='blue')
plt.title("Número de Registres d'Incidències per Àrees de l'Ajuntament de Barcelona")
plt.xlabel('Àrea')
plt.ylabel('Num. de Registres')

plt.xticks(rotation=90) #En aquest cas he posat 90 graus perquè són moltes les variables i no es llegien bé perquè es solapaven.

plt.show()
```
